# SQLALCHEMY

Recommended approach.

```python
data_folderpath = './data/reservation_dataset.db'
engine = create_engine(f'sqlite:///{data_folderpath}')

# Write data from a DataFrame to SQLite using SQLAlchemy (the table is created if it doesn't exist)
# df.to_sql('my_table', engine, if_exists='replace', index=False)

# Read data from SQLite into a DataFrame using SQLAlchemy
# df = pd.read_sql_table('my_table', engine)
```

```python
# Define the directory containing Parquet files
parquet_dir = Path('./data/')

# Loop through all Parquet files in the directory
for parquet_file in enumerate(parquet_dir.glob('*.parquet')):
    
    # Read the Parquet file into a DataFrame
    df = pd.read_parquet(parquet_file[1])
    df['Hotel_ID'] = parquet_file[0]
    
    table_name = 'reservations'
    
    df.to_sql(table_name, engine, if_exists='append', index=False)
```

```python
pd.read_sql_table('reservations', engine, dtype_backend="pyarrow")
```
